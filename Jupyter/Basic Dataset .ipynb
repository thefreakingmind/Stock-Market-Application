## Basic Of Machine Learning With Quandl for Financial Understanding
* Quandl API to collect Dataset of Stock Data

```python
import pandas as pd
import quandl
from matplotlib import pyplot as plt
```

```python
df = quandl.get('WIKI/GOOGL')
```

```python
df
```

```python
df[['Adj. Open', 'Adj. Close', 'Adj. High', 'Adj. Low', 'Adj. Volume']]
```

```python
plt.plot(df.head())
plt.title('Stock Market')
```

```python
df['HLT'] = df['Adj. Open'] - df['Adj. Close']/df['Adj. Close'] * 100
df['PCT_CHANGE'] = df['Adj. High'] - df['Adj. Low']/df['Adj. Low'] * 100
```

```python
plt.plot(df['HLT'])
```
